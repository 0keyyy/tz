# 2

1. Последний клик (Last Click):
   
   Преимущества: прост в реализаии и понимании, может быть эффективен для прямых маркетинговых каналов

   Критические ограничения: не учитывает другие маркетинговые каналы, которые могли также влиять на установку, не учитывает долгосрочные воздействия маркетинговых компаний
   

2. Линейная модель (Linear model):

   Преимущества: учитывает вклад всех маркетинговых каналов и действий, подходит для проектов с несколькими каналами привлечения

   Критические ограничения: может недооценить важность ключевых маркетинговых каналов, не учитывает временные различия во влиянии каналов
   

3. Атрибуция по времени (Time Decay):

   Преимущества: учитывает временную близость маркетинговых каналов к установке, позволяет учитывать долгосрочные и краткосрочные воздействия

   Критические ограничения: может недооценить ранние этапы привлечения, может быть солжно определить оптимальный временной интервал для учета
   

4. Модель атрибуции на основе данных (Data-Driven Attribution):

   Преимущества: использует данные о поведении пользователей для определения влияния маркетинговых каналов, может учитывать индивидуальные особенности проекта

   Критические ограничения: требует доступа к большому объему данных для точных результатов, может быть слоэно реализовать без специализированных инструментов и экспертизи в аналитике данных

Задача подразумевает случайный процесс с изменениями состояния и занимает область теории вероятностей и статистики. Поэтому точное решение может быть весьма сложным (постучимся к методу Монте-Карло). Прохождение уровня занимаетет 3м или 180с. Вероятность проигрыша при любой попытке составляет 75%, что приводит к тому, что на одну успешную попытку приходится в среднем 4 попытки (1/(1-0,75)). Таким образом, среднее время, затраченное на прохождение одного уровня, составляет 720с. Также после каждой попытки (будь то успешное прохождение уровня или поражение), игрок может взять паузу 10-30с с вероятностью 20% Предположим, что среднее время перерыва составляет 20 секунд. Тогда среднее время паузы после каждой игры составит 4с(20% от 20с). Таким образом, полное время, затрачиваемое на одну попытку (включая паузу), составляет 720 + 4 = 724с. В одном часе 3600 секунд, таким образом, предположительное количество уровней, которые игрок пройдет за один час, составит 3600 / 724 ≈ 4,97, это скорее всего будет округлено до 5 уровней. Это предположительные и усредненные значения. Реальное количество уровней может значительно варьироваться в зависимости от специфичных обстоятельств каждой игры.
